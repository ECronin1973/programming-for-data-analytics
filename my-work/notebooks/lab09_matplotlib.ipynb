### set()

Setting parametes on plot using

set()

it can take parameters xlim, xlabel, title and many more

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set(xlabel="x axis label", ylabel="y", title="hey mom")

plt.show()
```

### The axis functions
Or we can use the functions in axis

set_title set_xlabel set_ylabel

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

plt.show()
```

### Format the labels
Use a formatter function to format the labels using

ax.xaxis.set_major_formatter(formatter_fun)

The function that will do the formatting
It takes two parameters the value on the axis and the position of the tick It should return text as it should appear

```python
def currency_formatter(x,pos):
    s = f"${x}M"
    return s
```

Pass the function to

ax.xaxis.set_major_formatter()

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```

### Add text
Using text()

ax.text(x, y, "text to add", fontsize=10, verticalalignment="center")

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

ax.text(1, 8, "blah", fontsize=10, verticalalignment="center")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

### Annotations
I love these

ax.annotate('text', (x_pointer, y_pointer), xytext=(x, y)) 

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

#ax.text(1, 8, "blah", fontsize=10, verticalalignment="center")
ax.annotate('max', (3.2, 8), xytext=(1, 9),arrowprops=dict(facecolor='black', shrink=0.05))

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```
